```python
import numpy as np
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import torch
from torch.utils.data import Dataset, DataLoader, random_split
from torch import nn
import torch.nn.functional as F
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from torch.cuda.amp import autocast, GradScaler
from torchmetrics.classification import MultilabelF1Score

# Install ESM if not available
try:
    import esm
except ImportError:
    print("ESM not found. Installing...")
    import subprocess
    import sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "fair-esm"])
    import esm
    print("✓ ESM installed successfully!")

MAIN_DIR = '/kaggle/input/cafa-6-protein-function-prediction'

class Config:
    train_sequences_path = MAIN_DIR + "/Train/train_sequences.fasta"
    train_labels_path = MAIN_DIR + "/Train/train_terms.tsv"
    test_sequences_path = MAIN_DIR + "/Test/testsuperset.fasta"
    ia_path = MAIN_DIR + "/IA.tsv"
    
    num_labels = 500
    n_epochs = 10
    batch_size = 16  # Reduced for fine-tuning with GPU
    accumulation_steps = 4  # Increased to compensate
    lr = 1e-5  # Lower LR for fine-tuning
    weight_decay = 0.01
    warmup_steps = 1000
    max_grad_norm = 1.0
    label_smoothing = 0.1
    focal_alpha = 0.25
    focal_gamma = 2.0
    
    # ESM model settings
    esm_model_name = "esm2_t33_650M_UR50D"  # Options: esm2_t6_8M, esm2_t12_35M, esm2_t30_150M, esm2_t33_650M
    freeze_esm_layers = 20  # Freeze first N layers, fine-tune the rest
    use_pretrained_embeddings = False  # Set True to use pre-computed embeddings only
    
    # Model architecture
    d_model = 1280  # ESM2 650M output dim (adjust based on model: 320/480/640/1280)
    num_heads = 16
    num_layers = 4
    dropout = 0.2
    
    # GPU optimization settings
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    use_amp = True  # Mixed precision training
    num_workers = 4  # DataLoader workers
    pin_memory = True  # Pin memory for faster GPU transfer
    persistent_workers = True  # Keep workers alive between epochs
    prefetch_factor = 2  # Prefetch batches
    
    # Multi-GPU settings
    use_ddp = False  # Distributed Data Parallel
    
    @classmethod
    def print_gpu_info(cls):
        """Print GPU information"""
        if torch.cuda.is_available():
            print("\n" + "="*60)
            print("GPU INFORMATION")
            print("="*60)
            print(f"Device: {cls.device}")
            print(f"GPU Name: {torch.cuda.get_device_name(0)}")
            print(f"GPU Count: {torch.cuda.device_count()}")
            print(f"CUDA Version: {torch.version.cuda}")
            print(f"cuDNN Version: {torch.backends.cudnn.version()}")
            print(f"Memory Allocated: {torch.cuda.memory_allocated(0) / 1024**3:.2f} GB")
            print(f"Memory Reserved: {torch.cuda.memory_reserved(0) / 1024**3:.2f} GB")
            print(f"Max Memory Allocated: {torch.cuda.max_memory_allocated(0) / 1024**3:.2f} GB")
            print(f"Mixed Precision (AMP): {cls.use_amp}")
            print("="*60 + "\n")
        else:
            print("\n⚠ WARNING: No GPU available! Running on CPU (will be slow)")
            print("Consider using a GPU runtime for better performance\n")


class ProteinSequenceDataset(Dataset):
    """Dataset that loads sequences directly for ESM processing"""
    def __init__(self, sequences_path, labels_path=None, ia_weights=None, datatype="train"):
        super().__init__()
        self.datatype = datatype
        
        print(f"Loading sequences from: {sequences_path}")
        
        # Check if file exists
        import os
        if not os.path.exists(sequences_path):
            raise FileNotFoundError(f"Sequence file not found: {sequences_path}")
        
        # Load sequences from FASTA
        self.sequences = []
        self.ids = []
        
        try:
            with open(sequences_path, 'r') as f:
                current_id = None
                current_seq = []
                
                for line in f:
                    line = line.strip()
                    if line.startswith('>'):
                        if current_id is not None:
                            self.ids.append(current_id)
                            self.sequences.append(''.join(current_seq))
                        current_id = line[1:].split()[0]  # Take first word after >
                        current_seq = []
                    else:
                        current_seq.append(line)
                
                # Add last sequence
                if current_id is not None:
                    self.ids.append(current_id)
                    self.sequences.append(''.join(current_seq))
            
            print(f"✓ Loaded {len(self.sequences)} sequences")
            if len(self.sequences) > 0:
                avg_len = np.mean([len(seq) for seq in self.sequences])
                print(f"  Average sequence length: {avg_len:.0f}")
        
        except Exception as e:
            raise RuntimeError(f"Error loading sequences: {str(e)}")
        
        # Load labels for training
        if datatype == "train" and labels_path is not None:
            print(f"Loading labels from: {labels_path}")
            
            if not os.path.exists(labels_path):
                raise FileNotFoundError(f"Labels file not found: {labels_path}")
            
            try:
                labels_df = pd.read_csv(labels_path, sep='\t', names=['EntryID', 'term', 'aspect'])
                
                # Get top 500 terms
                top_terms = labels_df.groupby('term').size().sort_values(ascending=False).head(Config.num_labels)
                self.top_terms = top_terms.index.tolist()
                
                print(f"✓ Using top {len(self.top_terms)} GO terms")
                
                # Create binary labels
                self.labels = {}
                for entry_id in self.ids:
                    entry_labels = labels_df[labels_df['EntryID'] == entry_id]['term'].values
                    label_vector = np.zeros(Config.num_labels, dtype=np.float32)
                    for i, term in enumerate(self.top_terms):
                        if term in entry_labels:
                            label_vector[i] = 1.0
                    self.labels[entry_id] = label_vector
                
                print(f"✓ Created labels for {len(self.labels)} proteins")
                
                # Load IA weights
                if ia_weights is not None:
                    self.ia_weights = torch.tensor(ia_weights, dtype=torch.float32)
                else:
                    self.ia_weights = torch.ones(Config.num_labels)
            
            except Exception as e:
                raise RuntimeError(f"Error loading labels: {str(e)}")
    
    def __len__(self):
        return len(self.sequences)
    
    def __getitem__(self, index):
        seq = self.sequences[index]
        entry_id = self.ids[index]
        
        if self.datatype == "train":
            targets = torch.tensor(self.labels[entry_id], dtype=torch.float32)
            return entry_id, seq, targets, self.ia_weights
        else:
            return entry_id, seq


class ProteinEmbeddingDataset(Dataset):
    """Dataset for pre-computed ESM embeddings"""
    def __init__(self, emb_path, id_path, labels_path=None, ia_weights=None, datatype="train"):
        super().__init__()
        self.datatype = datatype
        
        # Load embeddings
        embeds = np.load(emb_path)
        ids = np.load(id_path)
        
        # Data augmentation: add noise to embeddings
        if datatype == "train":
            noise = np.random.normal(0, 0.01, embeds.shape)
            embeds = embeds + noise
        
        self.df = pd.DataFrame({
            "EntryID": ids,
            "embed": [embeds[i] for i in range(embeds.shape[0])]
        })
        
        if datatype == "train":
            np_labels = np.load(labels_path)
            df_labels = pd.DataFrame(self.df['EntryID'])
            df_labels['labels_vect'] = [row for row in np_labels]
            self.df = self.df.merge(df_labels, on="EntryID")
            
            if ia_weights is not None:
                self.ia_weights = torch.tensor(ia_weights, dtype=torch.float32)
            else:
                self.ia_weights = torch.ones(Config.num_labels)
    
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, index):
        embed = torch.tensor(self.df.iloc[index]["embed"], dtype=torch.float32)
        
        if self.datatype == "train":
            targets = torch.tensor(self.df.iloc[index]["labels_vect"], dtype=torch.float32)
            return embed, targets, self.ia_weights
        else:
            id = self.df.iloc[index]["EntryID"]
            return embed, id


def collate_sequences(batch):
    """Collate function for sequence batches"""
    if len(batch[0]) == 4:  # Training
        ids, seqs, targets, weights = zip(*batch)
        return list(ids), list(seqs), torch.stack(targets), weights[0]
    else:  # Testing
        ids, seqs = zip(*batch)
        return list(ids), list(seqs)


class FocalLoss(nn.Module):
    """Focal Loss for handling class imbalance"""
    def __init__(self, alpha=0.25, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
    
    def forward(self, inputs, targets, weights=None):
        bce = F.binary_cross_entropy_with_logits(inputs, targets, reduction='none')
        pt = torch.exp(-bce)
        focal = self.alpha * (1 - pt) ** self.gamma * bce
        
        if weights is not None:
            focal = focal * weights.unsqueeze(0)
        
        return focal.mean()


class HierarchicalLoss(nn.Module):
    """Loss that considers GO hierarchy"""
    def __init__(self, go_graph=None):
        super().__init__()
        self.focal = FocalLoss(alpha=Config.focal_alpha, gamma=Config.focal_gamma)
        self.go_graph = go_graph
    
    def forward(self, preds, targets, weights=None):
        loss = self.focal(preds, targets, weights)
        
        if self.go_graph is not None:
            hierarchy_penalty = self._compute_hierarchy_penalty(preds)
            loss = loss + 0.1 * hierarchy_penalty
        
        return loss
    
    def _compute_hierarchy_penalty(self, preds):
        return torch.tensor(0.0, device=preds.device)


class ESMProteinFunctionPredictor(nn.Module):
    """Protein function predictor with ESM backbone"""
    def __init__(self, num_classes=500, fine_tune=True, input_dim=None):
        super().__init__()
        
        self.fine_tune = fine_tune
        
        # Load ESM model
        if fine_tune:
            print(f"Loading ESM model: {Config.esm_model_name}")
            print("This may take a few minutes on first run (downloading ~2.5GB)...")
            
            try:
                # Download and load ESM model with progress
                import urllib.request
                import sys
                
                class DownloadProgressBar():
                    def __init__(self):
                        self.pbar = None
                    
                    def __call__(self, block_num, block_size, total_size):
                        if not self.pbar:
                            self.pbar = tqdm(total=total_size, unit='B', unit_scale=True, desc='Downloading ESM')
                        downloaded = block_num * block_size
                        if downloaded < total_size:
                            self.pbar.update(block_size)
                        else:
                            self.pbar.close()
                
                # Set timeout for model loading
                import socket
                socket.setdefaulttimeout(300)  # 5 minutes timeout
                
                self.esm_model, self.alphabet = esm.pretrained.load_model_and_alphabet(
                    Config.esm_model_name
                )
                
                print("✓ ESM model loaded successfully!")
                
                self.batch_converter = self.alphabet.get_batch_converter()
                
                # Freeze early layers
                if Config.freeze_esm_layers > 0:
                    print(f"Freezing first {Config.freeze_esm_layers} ESM layers...")
                    layer_count = 0
                    for name, param in self.esm_model.named_parameters():
                        if 'layers' in name:
                            layer_idx = int(name.split('.')[1])
                            if layer_idx < Config.freeze_esm_layers:
                                param.requires_grad = False
                                layer_count += 1
                    print(f"✓ Froze {layer_count} layer parameters")
                
                # ESM embedding dimension
                esm_dim = self.esm_model.embed_dim
                print(f"ESM embedding dimension: {esm_dim}")
                
            except Exception as e:
                print(f"\n✗ Error loading ESM model: {str(e)}")
                print("This could be due to:")
                print("  1. Network timeout (try again)")
                print("  2. Insufficient memory")
                print("  3. Firewall blocking download")
                print("\nFalling back to embeddings-only mode...")
                self.fine_tune = False
                esm_dim = input_dim if input_dim else Config.d_model
                
        else:
            # Use provided input dimension or default
            esm_dim = input_dim if input_dim else Config.d_model
            print(f"Using embeddings-only mode (input dim: {esm_dim})")
        
        # Store input dimension
        self.input_dim = esm_dim
        
        # Projection layer
        self.projection = nn.Sequential(
            nn.Linear(esm_dim, Config.d_model),
            nn.LayerNorm(Config.d_model),
            nn.GELU(),
            nn.Dropout(Config.dropout)
        )
        
        # Attention pooling (better than mean pooling)
        self.attention_pool = nn.Sequential(
            nn.Linear(Config.d_model, 1),
            nn.Softmax(dim=1)
        )
        
        # Multi-scale feature extraction
        self.multi_scale = nn.ModuleList([
            nn.Sequential(
                nn.Conv1d(Config.d_model, Config.d_model // 4, kernel_size=k, padding=k//2),
                nn.BatchNorm1d(Config.d_model // 4),
                nn.GELU()
            )
            for k in [3, 5, 7, 9]
        ])
        
        # Global pooling
        self.pool = nn.AdaptiveAvgPool1d(1)
        
        # Classification head
        self.classifier = nn.Sequential(
            nn.Linear(Config.d_model * 2, Config.d_model),
            nn.LayerNorm(Config.d_model),
            nn.GELU(),
            nn.Dropout(0.3),
            nn.Linear(Config.d_model, Config.d_model // 2),
            nn.LayerNorm(Config.d_model // 2),
            nn.GELU(),
            nn.Dropout(0.2),
            nn.Linear(Config.d_model // 2, num_classes)
        )
        
        # Temperature parameter
        self.temperature = nn.Parameter(torch.ones(1) * 1.5)
    
    def forward(self, sequences=None, embeddings=None):
        """
        Forward pass - can accept either sequences or pre-computed embeddings
        Args:
            sequences: List of (id, sequence) tuples for ESM processing
            embeddings: Pre-computed embeddings tensor [batch, dim] or [batch, seq_len, dim]
        """
        if sequences is not None and self.fine_tune:
            # Process sequences through ESM
            _, _, batch_tokens = self.batch_converter(sequences)
            batch_tokens = batch_tokens.to(next(self.parameters()).device)
            
            # Get ESM representations
            with torch.amp.autocast('cuda', enabled=Config.use_amp):
                results = self.esm_model(batch_tokens, repr_layers=[33], return_contacts=False)
                x = results["representations"][33]  # [batch, seq_len, esm_dim]
        else:
            # Use pre-computed embeddings
            if embeddings is None:
                raise ValueError("Must provide either sequences or embeddings")
            x = embeddings
            
            # Handle both 2D [batch, dim] and 3D [batch, seq_len, dim] embeddings
            if x.dim() == 2:
                # Add sequence dimension: [batch, dim] -> [batch, 1, dim]
                x = x.unsqueeze(1)
        
        # Project to model dimension
        x = self.projection(x)  # [batch, seq_len, d_model]
        
        # Attention pooling
        attn_weights = self.attention_pool(x)  # [batch, seq_len, 1]
        x_attn = (x * attn_weights).sum(dim=1)  # [batch, d_model]
        
        # Multi-scale features
        x_permuted = x.permute(0, 2, 1)  # [batch, d_model, seq_len]
        multi_scale_feats = [conv(x_permuted) for conv in self.multi_scale]
        multi_scale_feats = torch.cat(multi_scale_feats, dim=1)
        multi_scale_pooled = self.pool(multi_scale_feats).squeeze(-1)
        
        # Concatenate features
        combined = torch.cat([x_attn, multi_scale_pooled], dim=1)
        
        # Classification
        logits = self.classifier(combined)
        
        return logits
    
    def predict_calibrated(self, sequences=None, embeddings=None):
        """Predictions with temperature scaling"""
        logits = self.forward(sequences, embeddings)
        return torch.sigmoid(logits / self.temperature)


def train_model_with_esm(model, train_dataset, val_dataset, use_sequences=True):
    """Training with ESM model - GPU optimized"""
    
    # GPU-optimized DataLoader settings
    if use_sequences:
        train_loader = DataLoader(
            train_dataset, 
            batch_size=Config.batch_size,
            shuffle=True, 
            collate_fn=collate_sequences,
            num_workers=Config.num_workers,
            pin_memory=Config.pin_memory,
            persistent_workers=Config.persistent_workers if Config.num_workers > 0 else False,
            prefetch_factor=Config.prefetch_factor if Config.num_workers > 0 else None
        )
        val_loader = DataLoader(
            val_dataset, 
            batch_size=Config.batch_size,
            shuffle=False, 
            collate_fn=collate_sequences,
            num_workers=Config.num_workers,
            pin_memory=Config.pin_memory,
            persistent_workers=Config.persistent_workers if Config.num_workers > 0 else False
        )
    else:
        train_loader = DataLoader(
            train_dataset, 
            batch_size=Config.batch_size,
            shuffle=True, 
            num_workers=Config.num_workers, 
            pin_memory=Config.pin_memory,
            persistent_workers=Config.persistent_workers if Config.num_workers > 0 else False,
            prefetch_factor=Config.prefetch_factor if Config.num_workers > 0 else None
        )
        val_loader = DataLoader(
            val_dataset, 
            batch_size=Config.batch_size,
            shuffle=False, 
            num_workers=Config.num_workers, 
            pin_memory=Config.pin_memory,
            persistent_workers=Config.persistent_workers if Config.num_workers > 0 else False
        )
    
    # Enable cuDNN autotuner for optimal performance
    torch.backends.cudnn.benchmark = True
    
    optimizer = torch.optim.AdamW(model.parameters(), lr=Config.lr,
                                  weight_decay=Config.weight_decay)
    scheduler = CosineAnnealingWarmRestarts(optimizer, T_0=10, T_mult=2)
    criterion = HierarchicalLoss()
    scaler = GradScaler(enabled=Config.use_amp)
    f1_metric = MultilabelF1Score(num_labels=Config.num_labels).to(Config.device)
    
    train_loss_history = []
    val_loss_history = []
    train_f1_history = []
    val_f1_history = []
    best_val_f1 = 0.0
    
    print("BEGIN ESM TRAINING...")
    print(f"Using device: {Config.device}")
    print(f"Mixed Precision: {Config.use_amp}")
    print(f"Batch size: {Config.batch_size}")
    print(f"Effective batch size: {Config.batch_size * Config.accumulation_steps}")
    
    for epoch in range(Config.n_epochs):
        print(f"\nEPOCH {epoch + 1}/{Config.n_epochs}")
        
        # Clear GPU cache at start of epoch
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
        
        # Training
        model.train()
        train_losses = []
        train_scores = []
        optimizer.zero_grad()
        
        for batch_idx, batch_data in enumerate(tqdm(train_loader, desc="Training")):
            if use_sequences:
                ids, seqs, targets, weights = batch_data
                sequences = [(id, seq) for id, seq in zip(ids, seqs)]
                
                targets = targets.to(Config.device, non_blocking=True)
                weights = weights.to(Config.device, non_blocking=True)
                
                # Label smoothing
                targets_smooth = targets * (1 - Config.label_smoothing) + Config.label_smoothing / 2
                
                with autocast(enabled=Config.use_amp):
                    preds = model(sequences=sequences)
                    loss = criterion(preds, targets_smooth, weights) / Config.accumulation_steps
                
                scaler.scale(loss).backward()
                
                if (batch_idx + 1) % Config.accumulation_steps == 0:
                    scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), Config.max_grad_norm)
                    scaler.step(optimizer)
                    scaler.update()
                    optimizer.zero_grad()
            else:
                embed, targets, weights = batch_data
                embed = embed.to(Config.device, non_blocking=True)
                targets = targets.to(Config.device, non_blocking=True)
                weights = weights.to(Config.device, non_blocking=True)
                
                targets_smooth = targets * (1 - Config.label_smoothing) + Config.label_smoothing / 2
                
                with autocast(enabled=Config.use_amp):
                    preds = model(embeddings=embed)
                    loss = criterion(preds, targets_smooth, weights) / Config.accumulation_steps
                
                scaler.scale(loss).backward()
                
                if (batch_idx + 1) % Config.accumulation_steps == 0:
                    scaler.unscale_(optimizer)
                    torch.nn.utils.clip_grad_norm_(model.parameters(), Config.max_grad_norm)
                    scaler.step(optimizer)
                    scaler.update()
                    optimizer.zero_grad()
            
            train_losses.append(loss.item() * Config.accumulation_steps)
            
            with torch.no_grad():
                score = f1_metric(preds, targets.int())
                train_scores.append(score.item())
        
        avg_train_loss = np.mean(train_losses)
        avg_train_f1 = np.mean(train_scores)
        train_loss_history.append(avg_train_loss)
        train_f1_history.append(avg_train_f1)
        
        # Print GPU memory usage
        if torch.cuda.is_available():
            print(f"GPU Memory: {torch.cuda.memory_allocated(0) / 1024**3:.2f} GB / {torch.cuda.max_memory_allocated(0) / 1024**3:.2f} GB")
        
        print(f"Train Loss: {avg_train_loss:.4f} | Train F1: {avg_train_f1:.4f}")
        
        # Validation
        model.eval()
        val_losses = []
        val_scores = []
        
        with torch.no_grad():
            for batch_data in tqdm(val_loader, desc="Validation"):
                if use_sequences:
                    ids, seqs, targets, weights = batch_data
                    sequences = [(id, seq) for id, seq in zip(ids, seqs)]
                    targets = targets.to(Config.device, non_blocking=True)
                    weights = weights.to(Config.device, non_blocking=True)
                    
                    with autocast(enabled=Config.use_amp):
                        preds = model(sequences=sequences)
                        loss = criterion(preds, targets, weights)
                else:
                    embed, targets, weights = batch_data
                    embed = embed.to(Config.device, non_blocking=True)
                    targets = targets.to(Config.device, non_blocking=True)
                    weights = weights.to(Config.device, non_blocking=True)
                    
                    with autocast(enabled=Config.use_amp):
                        preds = model(embeddings=embed)
                        loss = criterion(preds, targets, weights)
                
                val_losses.append(loss.item())
                score = f1_metric(preds, targets.int())
                val_scores.append(score.item())
        
        avg_val_loss = np.mean(val_losses)
        avg_val_f1 = np.mean(val_scores)
        val_loss_history.append(avg_val_loss)
        val_f1_history.append(avg_val_f1)
        
        print(f"Val Loss: {avg_val_loss:.4f} | Val F1: {avg_val_f1:.4f}")
        
        # Save best model
        if avg_val_f1 > best_val_f1:
            best_val_f1 = avg_val_f1
            torch.save(model.state_dict(), 'best_esm_model.pth')
            print(f"✓ New best model saved! F1: {best_val_f1:.4f}")
        
        scheduler.step()
        
        # Clear GPU cache after epoch
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
    
    print(f"\nTraining Complete! Best Val F1: {best_val_f1:.4f}")
    
    return model, {
        'train_loss': train_loss_history,
        'val_loss': val_loss_history,
        'train_f1': train_f1_history,
        'val_f1': val_f1_history
    }


# Example usage:
"""
# Option 1: Fine-tune ESM end-to-end (recommended for best performance)
train_dataset = ProteinSequenceDataset(
    Config.train_sequences_path,
    Config.train_labels_path,
    datatype="train"
)
model = ESMProteinFunctionPredictor(num_classes=Config.num_labels, fine_tune=True)
model = model.to(Config.device)
model, history = train_model_with_esm(model, train_set, val_set, use_sequences=True)

# Option 2: Use pre-computed ESM embeddings (faster, less memory)
train_dataset = ProteinEmbeddingDataset(
    train_emb_path, train_id_path, labels_path, datatype="train"
)
model = ESMProteinFunctionPredictor(num_classes=Config.num_labels, fine_tune=False)
model = model.to(Config.device)
model, history = train_model_with_esm(model, train_set, val_set, use_sequences=False)
"""

def load_ia_weights(ia_path, num_labels=500):
    """Load information accretion weights"""
    try:
        ia_df = pd.read_csv(ia_path, sep='\t', names=['term', 'weight'])
        weights = np.ones(num_labels)
        return weights
    except:
        return np.ones(num_labels)


def predict_with_tta(model, test_dataset, num_tta=3, use_sequences=True):
    """Prediction with test-time augmentation"""
    
    if use_sequences:
        test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False, 
                                collate_fn=collate_sequences)
    else:
        test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
    
    model.eval()
    
    # Load label names
    labels = pd.read_csv(Config.train_labels_path, sep="\t", names=['EntryID', 'term', 'aspect'])
    top_terms = labels.groupby("term")["EntryID"].count().sort_values(ascending=False)
    labels_names = top_terms[:Config.num_labels].index.values
    
    all_predictions = []
    all_ids = []
    
    print(f"Generating predictions with {num_tta} TTA iterations...")
    
    for tta_iter in range(num_tta):
        print(f"TTA iteration {tta_iter + 1}/{num_tta}")
        predictions = []
        ids_list = []
        
        with torch.no_grad():
            for batch_data in tqdm(test_loader):
                if use_sequences:
                    ids, seqs = batch_data
                    sequences = [(id, seq) for id, seq in zip(ids, seqs)]
                    
                    # Add slight noise for TTA (modify sequences slightly)
                    if tta_iter > 0:
                        # For sequences, we can use dropout during inference
                        model.train()  # Enable dropout
                        probs = model.predict_calibrated(sequences=sequences)
                        model.eval()
                    else:
                        probs = model.predict_calibrated(sequences=sequences)
                    
                    ids_list.extend(ids)
                else:
                    embed, id = batch_data
                    embed = embed.to(Config.device)
                    
                    # Add noise for TTA
                    if tta_iter > 0:
                        noise = torch.randn_like(embed) * 0.01
                        embed = embed + noise
                    
                    probs = model.predict_calibrated(embeddings=embed)
                    ids_list.append(id)
                
                predictions.append(probs.cpu().numpy())
        
        all_predictions.append(np.vstack(predictions))
        if tta_iter == 0:
            all_ids = ids_list
    
    # Average TTA predictions
    final_predictions = np.mean(all_predictions, axis=0)
    
    print(f"\nCreating submission file...")
    print(f"Number of test proteins: {len(all_ids)}")
    print(f"Number of GO terms: {Config.num_labels}")
    
    # Create submission DataFrame
    submission_rows = []
    for i, protein_id in enumerate(all_ids):
        for j, go_term in enumerate(labels_names):
            confidence = final_predictions[i, j]
            # Only include predictions above threshold
            if confidence > 0.01:
                submission_rows.append({
                    'Id': protein_id,
                    'GO term': go_term,
                    'Confidence': confidence
                })
    
    submission_df = pd.DataFrame(submission_rows)
    
    return submission_df


# MAIN EXECUTION
if __name__ == "__main__":
    import os
    
    print("="*60)
    print("ESM PROTEIN FUNCTION PREDICTION PIPELINE")
    print("="*60)
    
    # Print GPU info
    Config.print_gpu_info()
    
    # Optimize GPU settings
    if torch.cuda.is_available():
        # Enable TF32 for faster computation on Ampere GPUs
        torch.backends.cuda.matmul.allow_tf32 = True
        torch.backends.cudnn.allow_tf32 = True
        
        # Set memory management
        torch.cuda.empty_cache()
        
        print("GPU Optimizations Enabled:")
        print(f"  - cuDNN benchmark: True")
        print(f"  - TF32: {torch.backends.cuda.matmul.allow_tf32}")
        print(f"  - Mixed Precision (AMP): {Config.use_amp}")
        print()
    
    # Check if data files exist
    print("\nChecking data files...")
    train_seq_exists = os.path.exists(Config.train_sequences_path)
    train_labels_exists = os.path.exists(Config.train_labels_path)
    test_seq_exists = os.path.exists(Config.test_sequences_path)
    
    print(f"Train sequences: {'✓' if train_seq_exists else '✗'} {Config.train_sequences_path}")
    print(f"Train labels: {'✓' if train_labels_exists else '✗'} {Config.train_labels_path}")
    print(f"Test sequences: {'✓' if test_seq_exists else '✗'} {Config.test_sequences_path}")
    
    # FORCE EMBEDDINGS MODE for now (since files are likely missing)
    USE_FINE_TUNING = False
    
    # Determine which mode to use
    if train_seq_exists and train_labels_exists and False:  # Disabled for now
        USE_FINE_TUNING = True
        print("\n[Mode] Fine-tuning ESM end-to-end")
        print("Using raw sequence files for ESM fine-tuning\n")
    else:
        USE_FINE_TUNING = False
        print("\n[Mode] Using pre-computed embeddings")
        if not train_seq_exists:
            print("⚠ Sequence files not found, using embeddings mode\n")
        else:
            print("Using embeddings mode (faster and more stable)\n")
    
    # Load IA weights
    ia_weights = load_ia_weights(Config.ia_path, Config.num_labels)
    
    if USE_FINE_TUNING:
        try:
            # Create datasets from sequences
            print("Loading training data...")
            full_train_dataset = ProteinSequenceDataset(
                Config.train_sequences_path,
                Config.train_labels_path,
                ia_weights,
                datatype="train"
            )
            
            print("\nLoading test data...")
            test_dataset = ProteinSequenceDataset(
                Config.test_sequences_path,
                datatype="test"
            )
            
            # Split for training/validation
            train_size = int(0.9 * len(full_train_dataset))
            val_size = len(full_train_dataset) - train_size
            train_set, val_set = random_split(full_train_dataset, [train_size, val_size])
            
            print(f"\nDataset split:")
            print(f"  Training: {train_size}")
            print(f"  Validation: {val_size}")
            print(f"  Test: {len(test_dataset)}")
            
            # Initialize model with ESM backbone
            print(f"\nInitializing ESM model: {Config.esm_model_name}")
            model = ESMProteinFunctionPredictor(
                num_classes=Config.num_labels, 
                fine_tune=True
            )
            model = model.to(Config.device)
            
            # Print model info
            total_params = sum(p.numel() for p in model.parameters())
            trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
            print(f"\nModel Statistics:")
            print(f"  Total parameters: {total_params:,}")
            print(f"  Trainable parameters: {trainable_params:,}")
            print(f"  Frozen parameters: {total_params - trainable_params:,}")
            print(f"  Model size: ~{total_params * 4 / 1024**3:.2f} GB (FP32)")
            
            # Train
            print("\nStarting training...")
            model, history = train_model_with_esm(
                model, train_set, val_set, use_sequences=True
            )
            
            # Generate predictions
            print("\nGenerating test predictions...")
            submission_df = predict_with_tta(
                model, test_dataset, num_tta=3, use_sequences=True
            )
            
        except Exception as e:
            print(f"\n✗ Error in fine-tuning mode: {str(e)}")
            print("Falling back to embeddings mode...\n")
            USE_FINE_TUNING = False
    
    if not USE_FINE_TUNING:
        print("Using pre-computed embeddings mode")
        print("Please ensure you have pre-computed ESM embeddings or use the original ProtBERT embeddings\n")
        
        # Paths to pre-computed embeddings
        # Option 1: ESM embeddings (if you generated them)
        # dataset_path = "/kaggle/input/esm-embeddings"
        
        # Option 2: Original ProtBERT embeddings (fallback)
        dataset_path = "/kaggle/input/protbert-embeddings-for-cafa5"
        train_emb = f"{dataset_path}/train_embeddings.npy"
        train_id = f"{dataset_path}/train_ids.npy"
        test_emb = f"{dataset_path}/test_embeddings.npy"
        test_id = f"{dataset_path}/test_ids.npy"
        labels_path = "/kaggle/input/train-targets-top500/train_targets_top500.npy"
        
        # Check if embeddings exist
        print("Checking embedding files...")
        print(f"Train embeddings: {'✓' if os.path.exists(train_emb) else '✗'} {train_emb}")
        print(f"Train IDs: {'✓' if os.path.exists(train_id) else '✗'} {train_id}")
        print(f"Test embeddings: {'✓' if os.path.exists(test_emb) else '✗'} {test_emb}")
        print(f"Test IDs: {'✓' if os.path.exists(test_id) else '✗'} {test_id}")
        print(f"Labels: {'✓' if os.path.exists(labels_path) else '✗'} {labels_path}")
        
        if not all([os.path.exists(f) for f in [train_emb, train_id, test_emb, test_id, labels_path]]):
            print("\n✗ Required embedding files not found!")
            print("\nPlease either:")
            print("1. Provide the correct paths to your embedding files")
            print("2. Generate ESM embeddings first")
            print("3. Use the raw sequence files for fine-tuning")
            exit(1)
        
        try:
            # Create datasets
            print("\nLoading embedding datasets...")
            full_train_dataset = ProteinEmbeddingDataset(
                train_emb, train_id, labels_path, ia_weights, datatype="train"
            )
            
            test_dataset = ProteinEmbeddingDataset(
                test_emb, test_id, datatype="test"
            )
            
            # Split for training/validation
            train_size = int(0.9 * len(full_train_dataset))
            val_size = len(full_train_dataset) - train_size
            train_set, val_set = random_split(full_train_dataset, [train_size, val_size])
            
            print(f"\nDataset split:")
            print(f"  Training: {train_size}")
            print(f"  Validation: {val_size}")
            print(f"  Test: {len(test_dataset)}")
            
            # Initialize model without ESM backbone
            print("\nInitializing classification model...")
            
            # Detect embedding dimension from data
            sample_embed = full_train_dataset.df.iloc[0]['embed']
            if isinstance(sample_embed, np.ndarray):
                embedding_dim = sample_embed.shape[0]
            else:
                embedding_dim = len(sample_embed)
            
            print(f"Detected embedding dimension: {embedding_dim}")
            
            model = ESMProteinFunctionPredictor(
                num_classes=Config.num_labels, 
                fine_tune=False,
                input_dim=embedding_dim  # Use detected dimension
            )
            model = model.to(Config.device)
            
            total_params = sum(p.numel() for p in model.parameters())
            print(f"Model parameters: {total_params:,}")
            
            # Train
            print("\nStarting training...")
            model, history = train_model_with_esm(
                model, train_set, val_set, use_sequences=False
            )
            
            # Generate predictions
            print("\nGenerating test predictions...")
            submission_df = predict_with_tta(
                model, test_dataset, num_tta=3, use_sequences=False
            )
            
        except Exception as e:
            print(f"\n✗ Error: {str(e)}")
            import traceback
            traceback.print_exc()
            exit(1)
    
    # Plot training history
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    ax1.plot(history['train_loss'], label='Train Loss', marker='o')
    ax1.plot(history['val_loss'], label='Val Loss', marker='s')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.set_title('Training and Validation Loss')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    
    ax2.plot(history['train_f1'], label='Train F1', marker='o')
    ax2.plot(history['val_f1'], label='Val F1', marker='s')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('F1 Score')
    ax2.set_title('Training and Validation F1 Score')
    ax2.legend()
    ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('esm_training_history.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    print("\n" + "="*60)
    print("SUBMISSION FILE GENERATION")
    print("="*60)
    
    # Display submission statistics
    print(f"\n✓ Total predictions: {len(submission_df):,}")
    print(f"✓ Unique proteins: {submission_df['Id'].nunique():,}")
    print(f"✓ Unique GO terms: {submission_df['GO term'].nunique()}")
    print(f"✓ Average predictions per protein: {len(submission_df) / submission_df['Id'].nunique():.1f}")
    print(f"✓ Confidence range: [{submission_df['Confidence'].min():.4f}, {submission_df['Confidence'].max():.4f}]")
    print(f"✓ Mean confidence: {submission_df['Confidence'].mean():.4f}")
    
    # Show sample predictions
    print("\nSample predictions:")
    print(submission_df.head(10).to_string(index=False))
    
    # Save submission
    submission_df.to_csv('submission.tsv', sep='\t', header=False, index=False)
    print(f"\n✓ Submission file saved as 'submission.tsv'")
    
    # Optional: Save submission with headers for inspection
    submission_df.to_csv('submission_with_headers.csv', index=False)
    print(f"✓ Submission with headers saved as 'submission_with_headers.csv'")
    
    print("\n" + "="*60)
    print("PIPELINE COMPLETE!")
    print("="*60)
```
