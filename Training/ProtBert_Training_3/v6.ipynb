```python
import math
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import torch
from torch.utils.data import Dataset, DataLoader, random_split
from torch import nn
import torch.nn.functional as F
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts
from torch.cuda.amp import autocast, GradScaler
from torchmetrics.classification import MultilabelF1Score

class Config:
    MAIN_DIR = "/kaggle/input/cafa-6-protein-function-prediction"
    train_sequences_path = MAIN_DIR + "/Train/train_sequences.fasta"
    train_labels_path = MAIN_DIR + "/Train/train_terms.tsv"
    test_sequences_path = MAIN_DIR + "/Test/testsuperset.fasta"
    
    num_labels = 500
    n_epochs = 20
    batch_size = 32
    lr = 5e-4
    weight_decay = 1e-3
    grad_accumulation_steps = 2
    
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    seed = 42
    
    # Training enhancements
    use_focal_loss = True
    focal_alpha = 0.25
    focal_gamma = 2.0
    
    # Model architecture
    hidden_dims = [2048, 1024]
    num_transformer_layers = 4
    num_heads = 16
    dropout = 0.2

config = Config()

torch.manual_seed(config.seed)
np.random.seed(config.seed)

EMBEDS_MAP = {
    "T5": "t5embeds",
    "ProtBERT": "cafa6-protbert-embeddings",
    "ESM2": "cafa-6-ems-embedding-dataset"
}

EMBEDS_DIM = {
    "T5": 1024,
    "ProtBERT": 1024,
    "ESM2": 1280
}


def load_and_process_labels(labels_path, protein_ids, num_labels=500):
    """Load labels from TSV and create label matrix for top N GO terms"""
    print(f"\nLoading labels from: {labels_path}")
    
    # Load the TSV file
    labels_df = pd.read_csv(labels_path, sep="\t", header=None)
    labels_df.columns = ["EntryID", "term", "ontology"]
    
    # Get top N most frequent GO terms
    top_terms = labels_df.groupby("term")["EntryID"].count().sort_values(ascending=False)
    selected_terms = top_terms[:num_labels].index.values
    
    print(f"Selected top {num_labels} GO terms")
    print(f"Most common term appears {top_terms.iloc[0]} times")
    print(f"Least common selected term appears {top_terms.iloc[num_labels-1]} times")
    
    # Filter labels to only include selected terms
    labels_df = labels_df[labels_df["term"].isin(selected_terms)]
    
    # Create term to index mapping
    term_to_idx = {term: idx for idx, term in enumerate(selected_terms)}
    
    # Create label matrix
    label_matrix = np.zeros((len(protein_ids), num_labels), dtype=np.float32)
    
    # Create protein ID to index mapping
    id_to_idx = {pid: idx for idx, pid in enumerate(protein_ids)}
    
    # Fill label matrix
    for _, row in tqdm(labels_df.iterrows(), total=len(labels_df), desc="Processing labels"):
        protein_id = row["EntryID"]
        term = row["term"]
        
        if protein_id in id_to_idx and term in term_to_idx:
            protein_idx = id_to_idx[protein_id]
            term_idx = term_to_idx[term]
            label_matrix[protein_idx, term_idx] = 1.0
    
    # Print statistics
    labels_per_protein = label_matrix.sum(axis=1)
    print(f"\nLabel statistics:")
    print(f"  Mean labels per protein: {labels_per_protein.mean():.2f}")
    print(f"  Min labels per protein: {labels_per_protein.min():.0f}")
    print(f"  Max labels per protein: {labels_per_protein.max():.0f}")
    print(f"  Proteins with no labels: {(labels_per_protein == 0).sum()}")
    
    return label_matrix, selected_terms


class FocalLoss(nn.Module):
    """Focal Loss for handling class imbalance"""
    def __init__(self, alpha=0.25, gamma=2.0):
        super().__init__()
        self.alpha = alpha
        self.gamma = gamma
        
    def forward(self, inputs, targets):
        bce_loss = F.binary_cross_entropy_with_logits(inputs, targets, reduction='none')
        pt = torch.exp(-bce_loss)
        focal_loss = self.alpha * (1-pt)**self.gamma * bce_loss
        return focal_loss.mean()


class ProteinSequenceDataset(Dataset):    
    def __init__(self, datatype, embeddings_source, augment=False, label_matrix=None):
        super().__init__()
        self.datatype = datatype
        self.embeddings_source = embeddings_source
        self.augment = augment
        
        if embeddings_source in ["ProtBERT", "ESM2"]:
            embeds = np.load(f"/kaggle/input/{EMBEDS_MAP[embeddings_source]}/{datatype}_embeddings.npy")
            ids = np.load(f"/kaggle/input/{EMBEDS_MAP[embeddings_source]}/{datatype}_ids.npy")
        elif embeddings_source == "T5":
            embeds = np.load(f"/kaggle/input/{EMBEDS_MAP[embeddings_source]}/{datatype}_embeds.npy")
            ids = np.load(f"/kaggle/input/{EMBEDS_MAP[embeddings_source]}/{datatype}_ids.npy")
        
        self.df = pd.DataFrame({
            "EntryID": ids,
            "embed": [embeds[i] for i in range(embeds.shape[0])]
        })
        
        if datatype == "train":
            if label_matrix is None:
                raise ValueError("label_matrix must be provided for training data")
            
            # Add labels to dataframe
            df_labels = pd.DataFrame({
                'EntryID': ids,
                'labels_vect': [row for row in label_matrix]
            })
            self.df = self.df.merge(df_labels, on="EntryID")
            
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, index):
        embed = torch.tensor(self.df.iloc[index]["embed"], dtype=torch.float32)
        
        # Data augmentation for training
        if self.augment and self.datatype == "train":
            noise = torch.randn_like(embed) * 0.01
            embed = embed + noise
        
        if self.datatype == "train":
            targets = torch.tensor(self.df.iloc[index]["labels_vect"], dtype=torch.float32)
            return embed, targets
        else:
            protein_id = self.df.iloc[index]["EntryID"]
            return embed, protein_id


class TransformerBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, ff_dim, dropout=0.1):
        super().__init__()
        
        self.attention = nn.MultiheadAttention(
            embed_dim=embed_dim,
            num_heads=num_heads,
            dropout=dropout,
            batch_first=True
        )
        
        self.ffn = nn.Sequential(
            nn.Linear(embed_dim, ff_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(ff_dim, embed_dim),
            nn.Dropout(dropout)
        )
        
        self.norm1 = nn.LayerNorm(embed_dim)
        self.norm2 = nn.LayerNorm(embed_dim)
        
        self.dropout = nn.Dropout(dropout)
        
    def forward(self, x):
        attn_out, _ = self.attention(x, x, x)
        x = self.norm1(x + self.dropout(attn_out))
        
        ffn_out = self.ffn(x)
        x = self.norm2(x + ffn_out)
        
        return x


class PositionalEncoding(nn.Module):
    """Sinusoidal positional encoding for transformer."""
    def __init__(self, embed_dim, max_len=5000, dropout=0.1):
        super().__init__()
        self.dropout = nn.Dropout(p=dropout)
        
        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * (-math.log(10000.0) / embed_dim))
        
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        
        self.register_buffer('pe', pe)
        
    def forward(self, x):
        x = x + self.pe[:, :x.size(1), :]
        return self.dropout(x)


class ImprovedTransformerMLP(nn.Module):
    def __init__(self, input_dim, num_classes, 
                 hidden_dims=[2048, 1024], 
                 num_transformer_layers=4,
                 num_heads=16,
                 ff_expansion=4,
                 dropout=0.2):
        super().__init__()
        
        self.input_projection = nn.Linear(input_dim, hidden_dims[1])
        
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, hidden_dims[0]),
            nn.LayerNorm(hidden_dims[0]),
            nn.GELU(),
            nn.Dropout(dropout),
            
            nn.Linear(hidden_dims[0], hidden_dims[1]),
            nn.LayerNorm(hidden_dims[1]),
            nn.GELU(),
            nn.Dropout(dropout)
        )
        
        self.pos_encoding = PositionalEncoding(
            embed_dim=hidden_dims[1],
            dropout=0.1
        )
        
        self.transformer_blocks = nn.ModuleList([
            TransformerBlock(
                embed_dim=hidden_dims[1],
                num_heads=num_heads,
                ff_dim=hidden_dims[1] * ff_expansion,
                dropout=0.1
            )
            for _ in range(num_transformer_layers)
        ])
        
        self.final_norm = nn.LayerNorm(hidden_dims[1])
        
        self.decoder = nn.Sequential(
            nn.Linear(hidden_dims[1], 768),
            nn.LayerNorm(768),
            nn.GELU(),
            nn.Dropout(dropout),
            
            nn.Linear(768, 384),
            nn.LayerNorm(384),
            nn.GELU(),
            nn.Dropout(dropout),
            
            nn.Linear(384, num_classes)
        )
        
        self._init_weights()
        
    def _init_weights(self):
        for module in self.modules():
            if isinstance(module, nn.Linear):
                nn.init.kaiming_normal_(module.weight, mode='fan_out', nonlinearity='relu')
                if module.bias is not None:
                    nn.init.constant_(module.bias, 0)
            elif isinstance(module, (nn.LayerNorm, nn.BatchNorm1d)):
                nn.init.constant_(module.weight, 1)
                nn.init.constant_(module.bias, 0)
        
    def forward(self, x):
        residual = self.input_projection(x)
        
        encoded = self.encoder(x)
        encoded = encoded + residual
        
        x_seq = encoded.unsqueeze(1)
        x_seq = self.pos_encoding(x_seq)
        
        for transformer_block in self.transformer_blocks:
            x_seq = transformer_block(x_seq)
        x_seq = self.final_norm(x_seq)
        
        encoded = x_seq.squeeze(1)
        output = self.decoder(encoded)
        
        return output


def find_optimal_thresholds(model, val_loader, num_labels):
    print("\nFinding optimal thresholds...")
    model.eval()
    
    all_preds = []
    all_targets = []
    
    with torch.no_grad():
        for embed, targets in tqdm(val_loader, desc="Collecting validation predictions"):
            embed = embed.to(config.device)
            outputs = model(embed)
            preds = torch.sigmoid(outputs).cpu()
            
            all_preds.append(preds)
            all_targets.append(targets)
    
    all_preds = torch.cat(all_preds, dim=0)
    all_targets = torch.cat(all_targets, dim=0)
    
    thresholds = []
    for i in range(num_labels):
        best_f1 = 0
        best_thresh = 0.01
        
        for thresh in [0.01, 0.05, 0.1, 0.15, 0.2, 0.3, 0.4, 0.5]:
            preds_binary = (all_preds[:, i] > thresh).float()
            tp = (preds_binary * all_targets[:, i]).sum()
            fp = (preds_binary * (1 - all_targets[:, i])).sum()
            fn = ((1 - preds_binary) * all_targets[:, i]).sum()
            
            precision = tp / (tp + fp + 1e-8)
            recall = tp / (tp + fn + 1e-8)
            f1 = 2 * precision * recall / (precision + recall + 1e-8)
            
            if f1 > best_f1:
                best_f1 = f1
                best_thresh = thresh
        
        thresholds.append(best_thresh)
    
    return np.array(thresholds)


def train_model(embeddings_source, train_size=0.9):    
    print(f"\n{'='*60}")
    print(f"Training with {embeddings_source} embeddings")
    print(f"{'='*60}\n")
    
    # Load embeddings first to get protein IDs
    if embeddings_source in ["ProtBERT", "ESM2"]:
        train_ids = np.load(f"/kaggle/input/{EMBEDS_MAP[embeddings_source]}/train_ids.npy")
    elif embeddings_source == "T5":
        train_ids = np.load(f"/kaggle/input/{EMBEDS_MAP[embeddings_source]}/train_ids.npy")
    
    # Load and process labels
    label_matrix, go_terms = load_and_process_labels(
        config.train_labels_path, 
        train_ids, 
        config.num_labels
    )
    
    # Create dataset with processed labels
    train_dataset = ProteinSequenceDataset(
        datatype="train", 
        embeddings_source=embeddings_source,
        augment=True,
        label_matrix=label_matrix
    )
    
    train_len = int(len(train_dataset) * train_size)
    val_len = len(train_dataset) - train_len
    train_set, val_set = random_split(
        train_dataset,
        [train_len, val_len],
        generator=torch.Generator().manual_seed(config.seed)
    )
    
    train_loader = DataLoader(
        train_set,
        batch_size=config.batch_size,
        shuffle=True,
        num_workers=2,
        pin_memory=True
    )
    val_loader = DataLoader(
        val_set,
        batch_size=config.batch_size,
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )
    
    model = nn.DataParallel(
        ImprovedTransformerMLP(
            input_dim=EMBEDS_DIM[embeddings_source],
            num_classes=config.num_labels,
            hidden_dims=config.hidden_dims,
            num_transformer_layers=config.num_transformer_layers,
            num_heads=config.num_heads,
            ff_expansion=4,
            dropout=config.dropout
        )
    ).to(config.device)
    
    print(f"Input dim: {EMBEDS_DIM[embeddings_source]}, Output classes: {config.num_labels}")
    
    if config.use_focal_loss:
        criterion = FocalLoss(alpha=config.focal_alpha, gamma=config.focal_gamma)
        print("Using Focal Loss")
    else:
        criterion = nn.BCEWithLogitsLoss()
        print("Using BCE Loss")
    
    optimizer = torch.optim.AdamW(
        model.parameters(),
        lr=config.lr,
        weight_decay=config.weight_decay,
        betas=(0.9, 0.999)
    )
    
    scheduler = CosineAnnealingWarmRestarts(
        optimizer,
        T_0=5,
        T_mult=2,
        eta_min=1e-6
    )
    
    scaler = GradScaler()
    
    f1_metric = MultilabelF1Score(num_labels=config.num_labels, average='macro').to(config.device)
    
    history = {
        'train_loss': [],
        'val_loss': [],
        'train_f1': [],
        'val_f1': []
    }
    
    best_val_f1 = 0.0
    best_model_state = None
    patience_counter = 0
    early_stopping_patience = 7
    
    print("Starting training...")
    for epoch in range(config.n_epochs):
        print(f"\nEpoch {epoch+1}/{config.n_epochs}")
        print("-" * 40)
        
        model.train()
        train_losses, train_f1s = [], []
        
        optimizer.zero_grad()
        for batch_idx, (embed, targets) in enumerate(tqdm(train_loader, desc="Training")):
            embed = embed.to(config.device, non_blocking=True)
            targets = targets.to(config.device, non_blocking=True)
            
            with autocast():
                outputs = model(embed)
                loss = criterion(outputs, targets)
                loss = loss / config.grad_accumulation_steps
            
            scaler.scale(loss).backward()
            
            if (batch_idx + 1) % config.grad_accumulation_steps == 0:
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                scaler.step(optimizer)
                scaler.update()
                optimizer.zero_grad()
            
            train_losses.append(loss.item() * config.grad_accumulation_steps)
            
            with torch.no_grad():
                preds = torch.sigmoid(outputs)
                f1 = f1_metric(preds, targets.int())
                train_f1s.append(f1.item())
        
        avg_train_loss = np.mean(train_losses)
        avg_train_f1 = np.mean(train_f1s)
        
        model.eval()
        val_losses, val_f1s = [], []
        
        with torch.no_grad():
            for embed, targets in tqdm(val_loader, desc="Validation"):
                embed = embed.to(config.device, non_blocking=True)
                targets = targets.to(config.device, non_blocking=True)
                
                with autocast():
                    outputs = model(embed)
                    loss = criterion(outputs, targets)
                
                preds = torch.sigmoid(outputs)
                f1 = f1_metric(preds, targets.int())
                
                val_losses.append(loss.item())
                val_f1s.append(f1.item())
        
        avg_val_loss = np.mean(val_losses)
        avg_val_f1 = np.mean(val_f1s)
        
        history['train_loss'].append(avg_train_loss)
        history['val_loss'].append(avg_val_loss)
        history['train_f1'].append(avg_train_f1)
        history['val_f1'].append(avg_val_f1)
        
        print(f"Train Loss: {avg_train_loss:.4f} | Train F1: {avg_train_f1:.4f}")
        print(f"Val Loss:   {avg_val_loss:.4f} | Val F1:   {avg_val_f1:.4f}")
        print(f"LR: {optimizer.param_groups[0]['lr']:.6f}")
        
        if avg_val_f1 > best_val_f1:
            best_val_f1 = avg_val_f1
            best_model_state = model.state_dict().copy()
            patience_counter = 0
            print(f"✓ New best model! (Val F1: {best_val_f1:.4f})")
        else:
            patience_counter += 1
            print(f"No improvement ({patience_counter}/{early_stopping_patience})")
        
        scheduler.step()
        
        if patience_counter >= early_stopping_patience:
            print(f"\nEarly stopping triggered after {epoch+1} epochs")
            break
    
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
        print(f"\n{'='*60}")
        print(f"Training complete! Best Val F1: {best_val_f1:.4f}")
        print(f"{'='*60}\n")
    
    thresholds = find_optimal_thresholds(model, val_loader, config.num_labels)
    
    return model, history, thresholds, go_terms


class TemperatureScaling:
    def __init__(self, temperature=1.5):
        self.temperature = temperature
    
    def calibrate(self, logits):
        return logits / self.temperature
    
    def fit_temperature(self, model, val_loader, device):
        from scipy.optimize import minimize
        
        all_logits = []
        all_targets = []
        
        model.eval()
        with torch.no_grad():
            for embed, targets in val_loader:
                embed = embed.to(device)
                logits = model(embed)
                all_logits.append(logits.cpu())
                all_targets.append(targets)
        
        all_logits = torch.cat(all_logits, dim=0)
        all_targets = torch.cat(all_targets, dim=0)
        
        def temperature_loss(temp):
            temp = temp[0]
            scaled_logits = all_logits / temp
            probs = torch.sigmoid(scaled_logits)
            loss = torch.nn.functional.binary_cross_entropy(
                probs, all_targets, reduction='mean'
            )
            return loss.item()
        
        result = minimize(temperature_loss, [1.5], bounds=[(0.5, 5.0)], method='L-BFGS-B')
        self.temperature = result.x[0]
        print(f"Optimal temperature found: {self.temperature:.3f}")
        
        return self.temperature


def generate_predictions_improved(model, embeddings_source, go_terms, thresholds=None, 
                                   val_loader=None, config=None,
                                   use_tta=True, n_augmentations=5,
                                   use_calibration=True,
                                   use_topk=True, min_predictions=5, max_predictions=50):
    print("\n" + "="*60)
    print("GENERATING IMPROVED PREDICTIONS")
    print("="*60)
    print(f"TTA: {'Enabled' if use_tta else 'Disabled'} (n={n_augmentations if use_tta else 0})")
    print(f"Calibration: {'Enabled' if use_calibration else 'Disabled'}")
    print(f"Top-K Filtering: {'Enabled' if use_topk else 'Disabled'} (min={min_predictions})")
    print("="*60 + "\n")
    
    test_dataset = ProteinSequenceDataset(
        datatype="test", 
        embeddings_source=embeddings_source
    )
    test_loader = DataLoader(
        test_dataset, 
        batch_size=1, 
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )
    
    if thresholds is None:
        thresholds = np.full(config.num_labels, 0.01)
        print("Using default thresholds: 0.01 for all terms")
    else:
        print(f"Using optimized thresholds (mean: {np.mean(thresholds):.3f})")
    
    temperature_scaler = None
    if use_calibration and val_loader is not None:
        print("\nCalibrating probabilities...")
        temperature_scaler = TemperatureScaling()
        temperature_scaler.fit_temperature(model, val_loader, config.device)
    elif use_calibration:
        print("\nUsing default temperature scaling (T=1.5)")
        temperature_scaler = TemperatureScaling(temperature=1.5)
    
    model.eval()
    all_ids = []
    all_terms = []
    all_confs = []
    
    proteins_with_no_predictions = 0
    proteins_with_topk_boost = 0
    
    with torch.no_grad():
        for embed, protein_id in tqdm(test_loader, desc="Predicting"):
            protein_id = protein_id[0]
            
            if use_tta:
                aug_logits = []
                
                for aug_idx in range(n_augmentations):
                    if aug_idx == 0:
                        aug_embed = embed
                    else:
                        noise = torch.randn_like(embed) * 0.008
                        aug_embed = embed + noise
                    
                    aug_embed = aug_embed.to(config.device)
                    logits = model(aug_embed)
                    aug_logits.append(logits.cpu())
                
                avg_logits = torch.mean(torch.stack(aug_logits), dim=0)
            else:
                embed = embed.to(config.device)
                avg_logits = model(embed).cpu()
            
            if use_calibration and temperature_scaler is not None:
                avg_logits = temperature_scaler.calibrate(avg_logits)
            
            probs = torch.sigmoid(avg_logits).numpy().flatten()
            
            above_threshold_indices = []
            above_threshold_probs = []
            
            for term_idx, conf in enumerate(probs):
                if conf > thresholds[term_idx]:
                    above_threshold_indices.append(term_idx)
                    above_threshold_probs.append(conf)
            
            if use_topk:
                if len(above_threshold_indices) < min_predictions:
                    proteins_with_topk_boost += 1
                    
                    top_k_indices = np.argsort(probs)[-min_predictions:][::-1]
                    
                    for idx in top_k_indices:
                        if idx not in above_threshold_indices:
                            above_threshold_indices.append(idx)
                            above_threshold_probs.append(probs[idx])
                
                elif len(above_threshold_indices) > max_predictions:
                    sorted_pairs = sorted(
                        zip(above_threshold_indices, above_threshold_probs),
                        key=lambda x: x[1],
                        reverse=True
                    )
                    above_threshold_indices = [x[0] for x in sorted_pairs[:max_predictions]]
                    above_threshold_probs = [x[1] for x in sorted_pairs[:max_predictions]]
            
            if len(above_threshold_indices) == 0:
                proteins_with_no_predictions += 1
            
            for term_idx, conf in zip(above_threshold_indices, above_threshold_probs):
                all_ids.append(protein_id)
                all_terms.append(go_terms[term_idx])
                all_confs.append(round(float(conf), 3))
    
    submission_df = pd.DataFrame({
        "Id": all_ids,
        "GO_term": all_terms,
        "Confidence": all_confs
    })
    
    print("\n" + "="*60)
    print("PREDICTION STATISTICS")
    print("="*60)
    print(f"Total predictions: {len(submission_df):,}")
    print(f"Unique proteins: {submission_df['Id'].nunique():,}")
    print(f"Unique GO terms: {submission_df['GO_term'].nunique()}")
    print(f"Avg predictions per protein: {len(submission_df) / submission_df['Id'].nunique():.1f}")
    print(f"Proteins with no predictions: {proteins_with_no_predictions}")
    if use_topk:
        print(f"Proteins boosted with Top-K: {proteins_with_topk_boost}")
    print(f"Confidence score range: [{submission_df['Confidence'].min():.3f}, {submission_df['Confidence'].max():.3f}]")
    print(f"Mean confidence: {submission_df['Confidence'].mean():.3f}")
    print("="*60 + "\n")
    
    return submission_df


def plot_prediction_statistics(submission_df):
    """Visualize prediction statistics"""
    import matplotlib.pyplot as plt
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    predictions_per_protein = submission_df.groupby('Id').size()
    axes[0, 0].hist(predictions_per_protein, bins=50, edgecolor='black', alpha=0.7)
    axes[0, 0].set_xlabel('Number of Predictions')
    axes[0, 0].set_ylabel('Number of Proteins')
    axes[0, 0].set_title('Distribution of Predictions per Protein')
    axes[0, 0].axvline(predictions_per_protein.mean(), color='red', 
                       linestyle='--', label=f'Mean: {predictions_per_protein.mean():.1f}')
    axes[0, 0].legend()
    
    # 2. Confidence score distribution
    axes[0, 1].hist(submission_df['Confidence'], bins=50, edgecolor='black', alpha=0.7)
    axes[0, 1].set_xlabel('Confidence Score')
    axes[0, 1].set_ylabel('Frequency')
    axes[0, 1].set_title('Distribution of Confidence Scores')
    axes[0, 1].axvline(submission_df['Confidence'].mean(), color='red',
                       linestyle='--', label=f'Mean: {submission_df["Confidence"].mean():.3f}')
    axes[0, 1].legend()
    
    # 3. Top 20 most predicted GO terms
    top_terms = submission_df['GO_term'].value_counts().head(20)
    axes[1, 0].barh(range(len(top_terms)), top_terms.values)
    axes[1, 0].set_yticks(range(len(top_terms)))
    axes[1, 0].set_yticklabels(top_terms.index, fontsize=8)
    axes[1, 0].set_xlabel('Prediction Count')
    axes[1, 0].set_title('Top 20 Most Predicted GO Terms')
    axes[1, 0].invert_yaxis()
    
    # 4. Confidence vs Prediction Count
    go_term_stats = submission_df.groupby('GO_term').agg({
        'Confidence': 'mean',
        'Id': 'count'
    }).rename(columns={'Id': 'count'})
    
    axes[1, 1].scatter(go_term_stats['count'], go_term_stats['Confidence'], alpha=0.5)
    axes[1, 1].set_xlabel('Number of Predictions')
    axes[1, 1].set_ylabel('Mean Confidence')
    axes[1, 1].set_title('GO Term Frequency vs Mean Confidence')
    axes[1, 1].set_xscale('log')
    
    plt.tight_layout()
    plt.show()


# Main execution
if __name__ == "__main__":
    # Train model
    model, history, thresholds, go_terms = train_model(embeddings_source="ESM2")
    
    # Load labels and create validation loader for calibration
    # First, load the training IDs
    train_ids = np.load(f"/kaggle/input/{EMBEDS_MAP['ESM2']}/train_ids.npy")
    
    # Load and process labels
    label_matrix, _ = load_and_process_labels(
        config.train_labels_path, 
        train_ids, 
        config.num_labels
    )
    
    # Create dataset with labels
    train_dataset = ProteinSequenceDataset(
        datatype="train", 
        embeddings_source="ESM2",
        label_matrix=label_matrix  # Now providing the label_matrix
    )
    
    # Split into train and validation
    train_len = int(len(train_dataset) * 0.9)
    val_len = len(train_dataset) - train_len
    train_set, val_set = random_split(
        train_dataset,
        [train_len, val_len],
        generator=torch.Generator().manual_seed(config.seed)
    )
    
    # Create validation loader
    val_loader = DataLoader(
        val_set, 
        batch_size=config.batch_size, 
        shuffle=False,
        num_workers=2,
        pin_memory=True
    )
    
    # Generate improved predictions
    submission_df = generate_predictions_improved(
        model=model,
        embeddings_source="ESM2",
        go_terms=go_terms,
        thresholds=thresholds,
        val_loader=val_loader,
        config=config,
        use_tta=True,
        n_augmentations=5,
        use_calibration=True,
        use_topk=True,
        min_predictions=5,
        max_predictions=50
    )
    
    # Visualize statistics
    plot_prediction_statistics(submission_df)
    
    # Save submission
    submission_df.to_csv('submission_improved.tsv', sep='\t', header=False, index=False)
    print("\nSubmission saved to 'submission_improved.tsv'")
```
